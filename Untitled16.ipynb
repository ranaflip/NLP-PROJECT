```python
import streamlit as st
import re

st.set_page_config(page_title="WhatsApp Action Item Extractor", page_icon="üìã")

st.title("üìã WhatsApp Action Item Extractor")
st.write("Upload a WhatsApp chat `.txt` file and extract actionable tasks and reminders!")

uploaded_file = st.file_uploader("Upload WhatsApp Chat File", type=["txt"])

if uploaded_file:
    lines = uploaded_file.read().decode("utf-8").splitlines()

    pattern = r'\[(\d{1,2}/\d{1,2}/\d{2,4}),\s(\d{1,2}:\d{2}:\d{2})\]\s(.*?):\s(.*)'
    messages = []

    for line in lines:
        match = re.match(pattern, line)
        if match:
            _, _, _, message = match.groups()
            if "omitted" not in message.lower() and "deleted" not in message.lower():
                messages.append(message.strip())

    st.success(f"‚úÖ Total {len(messages)} messages processed!")

    keywords = ['submit', 'complete', 'attend', 'register', 'upload', 'join',
                'need to', 'should', 'must', 'required', 'ensure', 'reminder', 'mandatory', 'prepare']

    def extract_action_items(messages):
        action_items = []
        for msg in messages:
            msg_lower = msg.lower()
            if any(kw in msg_lower for kw in keywords):
                action_items.append(msg)
        return action_items

    actions = extract_action_items(messages)

    if actions:
        st.markdown("### üîç Extracted Action Items")
        for i, act in enumerate(actions, 1):
            st.markdown(f"{i}. {act}")
        st.download_button("üì• Download as Text File", "\n".join(actions), file_name="action_items.txt")
    else:
        st.warning("üòê No actionable messages found.")
```
